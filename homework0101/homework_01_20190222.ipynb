```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# global
plt.figure(figsize=(7,7))
plt.xlabel("longitude")
plt.ylabel("latitude")

a = 90
b = 90
r = 90

latitude = [-60, -45, -23.5, 0, 23.5, 45, 60]
longitude = [-90, -60, -30, 0, 30, 60, 90]

def drawLat(latitude, degree):
    for lat in latitude:
        theta = np.linspace(0, degree, 1000)
        new_a = np.cos(np.pi * lat/180) * r
        rate = new_a/a
        new_b = b/8 * rate
        offset = np.sin(np.pi * lat/180) * r
        x = new_a * np.cos(theta)
        y = new_b * np.sin(theta) + offset
        if(degree < 0):
            plt.plot(x,y,color='grey',alpha=0.8) 
        else:
            plt.plot(x,y,'--', color='grey',alpha=0.5)
            
def drawLong(longitude):
    a = 90
    for long in longitude:  
        theta = np.linspace(np.pi/2, np.pi*3/2, 1000)
        x = a * np.cos(theta)
        y = b * np.sin(theta)
        plt.plot(x,y,color='gray')
        a -= r*2/(len(longitude)-1)
            
drawLat(latitude, -np.pi)
drawLat(latitude, np.pi)
drawLong(longitude)
```
