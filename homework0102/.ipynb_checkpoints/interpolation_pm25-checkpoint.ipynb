使用PM2.5的資料進行空間推估
===

### 資料來源
* [村里界圖(TWD97_121分帶)](https://data.gov.tw/dataset/7440)
* [pm2.5感測器資料](https://pm25.lass-net.org/GIS/IDW/data/data.json)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
import requests
import json
from shapely.geometry import Polygon, Point
```

### 將村里圖層合併為縣市圖層

```python
# read shapefile
filename = './data/taiwan_village_twd97.shp'
village = gpd.read_file(filename)

# town
town = village.dissolve(by='TOWNID')
town.to_file('./data/taiwan_town_twd97.shp', driver='ESRI Shapefile', encoding = 'utf-8')

# county
county = town.dissolve(by='COUNTYID')
county.to_file('./data/taiwan_county_twd97.shp', driver='ESRI Shapefile', encoding = 'utf-8')


county.plot(figsize=(6,6))
plt.show()
```

### 選出台北市作為範例

```python
taiwan_village = gpd.read_file('./data/taiwan_village_twd97.shp')
taiwan_town = gpd.read_file('./data/taiwan_town_twd97.shp')
taiwan_county = gpd.read_file('./data/taiwan_county_twd97.shp')

# village
taipei_village_twd97 = taiwan_village.loc[taiwan_village['COUNTYNAME'] == '臺北市']
taipei_village_twd97.to_file('./data/taipei_village_twd97.shp', driver = "ESRI Shapefile", encoding = 'utf-8')

# town
taipei_town_twd97 = taiwan_town.loc[taiwan_town['COUNTYNAME'] == '臺北市']
taipei_town_twd97.to_file('./data/taipei_town_twd97.shp', driver = "ESRI Shapefile", encoding = 'utf-8')

# county
taipei_county_twd97 = taiwan_county.loc[taiwan_county['COUNTYNAME'] == '臺北市']
taipei_county_twd97.to_file('./data/taipei_county_twd97.shp', driver = "ESRI Shapefile", encoding = 'utf-8')
```

### 台北市範圍展示

```python
taipei_village_twd97 = gpd.read_file('./data/taipei_village_twd97.shp')
taipei_town_twd97 = gpd.read_file('./data/taipei_town_twd97.shp')
taipei_county_twd97 = gpd.read_file('./data/taipei_county_twd97.shp')

county = taipei_county_twd97.plot(figsize=(6, 6), linewidth=2, facecolor='none', edgecolor='black', alpha=1)
town = taipei_town_twd97.plot(ax=county, facecolor='none', linewidth=1, edgecolor='black', alpha=0.5)
village = taipei_village_twd97.plot(ax=town, facecolor='none', linewidth=0.3, edgecolor='red', alpha=0.3)

plt.show()
```

### 抓pm2.5資料
* 座標為經緯度
* 為全世界的資料

```python
# get pm25 data
url = 'https://pm25.lass-net.org/GIS/IDW/data/data.json'
r = requests.get(url)
result = r.json()

# field
time = result['latest-updated-time']
points = result['points']
point_meta = result['point-meta']

pm25 = pd.DataFrame(points, columns=point_meta)
print(pm25.head())
```

### 選擇出台北市範圍內的感測器
* 需要轉換座標，座標資訊可參考[TWD97 / TM2 zone 121](https://epsg.io/3826)

```python
# WGS84 bounds:
# 114.32 17.36
# 123.61 26.96
pm25 = pm25.loc[pm25['gps_lat'] > 17.36]
pm25 = pm25.loc[pm25['gps_lat'] < 26.96]
pm25 = pm25.loc[pm25['gps_lon'] > 114.32]
pm25 = pm25.loc[pm25['gps_lon'] < 123.61]

# create geometry
pm25['Coordinates'] = list(zip(pm25.gps_lon, pm25.gps_lat))
pm25['Coordinates'] = pm25['Coordinates'].apply(Point)

# create geodataframe
pm25_point = gpd.GeoDataFrame(pm25, geometry='Coordinates')
pm25_point.crs = {'init' :'epsg:4326'}

# reproject
crs = "+proj=tmerc +lat_0=0 +lon_0=121 +k=0.9999 +x_0=250000 +y_0=0 +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs"
pm25_point = pm25_point.to_crs(crs)

# select layer by location (taipei county - twd97)
taipei_county_twd97 = gpd.read_file('./data/taipei_county_twd97.shp')

# geometry of taipei county (twd97)
polygon = taipei_county_twd97.geometry[0]
result = pm25_point.within(polygon)
result = result * 1

# clip
pm25['status'] = result.tolist()
pm25_point = gpd.GeoDataFrame(pm25, geometry='Coordinates')
pm25_point.crs = {'init' :'epsg:4326'}

# reproject
crs = "+proj=tmerc +lat_0=0 +lon_0=121 +k=0.9999 +x_0=250000 +y_0=0 +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs"
pm25_point = pm25_point.to_crs(crs)

taipei_pm25_point = pm25_point.loc[pm25_point['status'] == 1]

# output file 
taipei_pm25_point.to_file('./data/taipei_pm25_point.shp', driver='ESRI Shapefile', encoding='utf-8')

# display
taipei_county_twd97 = taipei_county_twd97.plot(figsize=(6,6), facecolor='none', edgecolor='black')
taipei_pm25_point.plot(ax=taipei_county_twd97)
plt.show()
```

### 空間推估
* [IDW](https://sourceforge.net/p/geoms2/wiki/Inverse%20weighted%20distance/)
* 依照圖層範圍產生網格

```python
def iwd(x,y,v,grid,power):
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            distance = np.sqrt((x-i)**2+(y-j)**2)
            if (distance**power).min()==0: 
                grid[i,j] = v[(distance**power).argmin()]
            else:
                total = np.sum(1/(distance**power))
                grid[i,j] = np.sum(v/(distance**power)/total)
    return grid

taipei_county_twd97 = gpd.read_file('./data/taipei_county_twd97.shp')
taipei_pm25_point = gpd.read_file('./data/taipei_pm25_point.shp')

# layer's bounds
bounds = taipei_county_twd97.bounds
minx = np.min(bounds['minx'])
miny = np.min(bounds['miny'])
maxx = np.max(bounds['maxx'])
maxy = np.max(bounds['maxy'])

# grid
size = 100
gridx = np.arange(minx, maxx, size)
gridy = np.arange(miny, maxy, size)

rows = len(gridy)
columns = len(gridx)

# type is Point
geometry = taipei_pm25_point.geometry

# get lat, lng and value
x = [int(np.ceil((lat - minx)/100)) for lat in geometry.x]
x = np.asarray(x)
y = [int(np.ceil((lng - miny)/100)) for lng in geometry.y]
y = np.asarray(y)
v = [value for value in taipei_pm25_point['pm2.5']]
v = np.asarray(v)

# rows and columns from layer's extent
grid = np.zeros((rows, columns),dtype='float32')
grid = iwd(y,x,v,grid,2)

# x, y origin
XleftOrigin = minx
XrightOrigin = minx + size
YtopOrigin = maxy
YbottomOrigin = maxy- size

polygons = []

for i in range(rows):
    Xleft = XleftOrigin
    Xright = XrightOrigin
    for j in range(columns):
        polygons.append(Polygon([(Xleft, YtopOrigin), (Xright, YtopOrigin), (Xright, YbottomOrigin), (Xleft, YbottomOrigin)]))
        Xleft = Xleft + size
        Xright = Xright + size        
    YtopOrigin = YtopOrigin - size
    YbottomOrigin = YbottomOrigin - size

grid = grid[::-1]
pm25_value = grid.ravel()

df = pd.DataFrame({
    'value': pm25_value
})

crs = "+proj=tmerc +lat_0=0 +lon_0=121 +k=0.9999 +x_0=250000 +y_0=0 +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs"
grid = gpd.GeoDataFrame(df, geometry=polygons, crs=crs)
grid.plot(column='value', cmap='RdYlGn_r')
plt.show()
grid.to_file("./data/grid_100m.shp")
```

### 裁切出台北市範圍的網格資料

```python
taipei_county_twd97 = gpd.read_file('./data/taipei_county_twd97.shp')
grid_100m = gpd.read_file('./data/grid_100m.shp')

intersection = gpd.overlay(taipei_county_twd97, grid_100m, how='intersection')
intersection.to_file('./data/taipei_grid_100m.shp', driver='ESRI Shapefile', encoding='utf-8')
```

### 展示

```python
taipei_village_twd97 = gpd.read_file('./data/taipei_village_twd97.shp')
taipei_town_twd97 = gpd.read_file('./data/taipei_town_twd97.shp')
taipei_county_twd97 = gpd.read_file('./data/taipei_county_twd97.shp')
taipei_pm25_point = gpd.read_file('./data/taipei_pm25_point.shp')
taipei_grid_100m = gpd.read_file('./data/taipei_grid_100m.shp')


county = taipei_county_twd97.plot(figsize=(10, 10), linewidth=1, facecolor='none', edgecolor='black', alpha=1)
town = taipei_town_twd97.plot(ax=county, facecolor='none', linewidth=1, edgecolor='black', alpha=0.5)
village = taipei_village_twd97.plot(ax=town, facecolor='none', linewidth=0.3, edgecolor='red', alpha=0.3)
point = taipei_pm25_point.plot(ax=village, markersize=3, label='sites')
grid = taipei_grid_100m.plot(ax=point, column='value', cmap='RdYlGn_r', alpha=0.3, legend=True)

plt.legend(loc='best')

plt.show()
```
